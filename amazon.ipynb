```python
### Import Libraries
import requests as rq
import pandas as pd
import json
from bs4 import BeautifulSoup 
import numpy as np
```

```python
url = 'https://www.amazon.in/s?k=bags&crid=2M096C61O4MLT&qid=1653308124&sprefix=ba%2Caps%2C283&ref=sr_pg_1'
dataframe=[]
headers =({"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:109.0) Gecko/20100101 Firefox/111.0",
           "Accept-Language":"en-US,en;q=0.5",
           })
for i in range(20):
    re=rq.get(url,headers=headers)
    print(re.status_code)
    page=BeautifulSoup(re.text,'html.parser')
    li=page.find_all(class_ = 's-card-container s-overflow-hidden aok-relative puis-include-content-margin puis s-latency-cf-section s-card-border')
    for k in li:
        data=[]
        review=k.find('span', class_ = 'a-size-base s-underline-text' )
        rating=k.find('span',class_='a-size-base')
        price=k.find('span',class_='a-price-whole')
        try:
            data.append('https://www.amazon.in/'+k.a['href'])
            data.append(k.h2.text)
            data.append(price.text.replace(',',''))
            data.append(rating.text)
            data.append(review.text[1:-1])
            dataframe.append(data)
        except AttributeError :
            continue
    next_page = page.find('a',class_='s-pagination-item s-pagination-next s-pagination-button s-pagination-separator')
    try:
        url='https://www.amazon.in'+next_page['href']
    except TypeError:
        break
```
